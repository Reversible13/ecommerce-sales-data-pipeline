```python
import pandas as pd
import glob
import os

# --- KONFIGURASI ---
folder_name = "Dataset Tugas Data Science M2"
path_pattern = os.path.join(folder_name, "*.xlsx")
all_files = glob.glob(path_pattern)

# Kolom standar yang akan digunakan
common_columns = [
    'order_id', 'product_category', 'Status Pesanan', 'Alasan Pembatalan', 
    'Opsi Pengiriman', 'Waktu Pesanan Dibuat', 'Metode Pembayaran', 'Jumlah', 
    'Returned quantity', 'Total Diskon', 'Total Berat', 'Ongkos Kirim Dibayar oleh Pembeli', 
    'Estimasi Potongan Biaya Pengiriman', 'Total Pembayaran', 'Perkiraan Ongkos Kirim', 
    'Kota/Kabupaten', 'Provinsi'
]

list_df = []

# --- TAHAP 1: DATA EXPLORATION & MERGING ---
if len(all_files) == 0:
    print(f"Peringatan: Tidak ada file .xlsx di folder '{folder_name}'")
else:
    print(f"Memulai Tahap 1: Memproses {len(all_files)} file...")
    for filename in all_files:
        try:
            df = pd.read_excel(filename)
            
            # Standardisasi nama kolom waktu
            if 'Waktu Pengiriman Diatur' in df.columns and 'Waktu Pesanan Dibuat' not in df.columns:
                df.rename(columns={'Waktu Pengiriman Diatur': 'Waktu Pesanan Dibuat'}, inplace=True)
            
            # Filter kolom standar
            cols_to_keep = [c for c in common_columns if c in df.columns]
            list_df.append(df[cols_to_keep].copy())
            
        except Exception as e:
            print(f"Gagal membaca {os.path.basename(filename)}: {e}")

    # Menggabungkan semua file menjadi df_master
    df_master = pd.concat(list_df, ignore_index=True)
    rows_initial = len(df_master)
    print("Tahap 1 Selesai: Semua file telah digabungkan.")

    # --- TAHAP 2: DATA CLEANSING ---
    print("\nMemulai Tahap 2: Data Cleansing...")

    # 1. Handling Missing Values
    df_master['Alasan Pembatalan'] = df_master['Alasan Pembatalan'].fillna("Tidak Batal")

    # 2. String Cleaning & Formatting
    # Bersihkan " gr" dan ubah ke float pada Total Berat
    if df_master['Total Berat'].dtype == 'object':
        df_master['Total Berat'] = df_master['Total Berat'].str.replace(' gr', '', regex=False).astype(float)
    
    # Ubah ke format Datetime
    df_master['Waktu Pesanan Dibuat'] = pd.to_datetime(df_master['Waktu Pesanan Dibuat'])

    # 3. Handling Duplicates
    df_master = df_master.drop_duplicates()
    rows_final = len(df_master)

    # --- LAPORAN AKHIR ---
    print("="*50)
    print("PROSES INTEGRASI BERHASIL")
    print("="*50)
    print(f"Total Baris Awal (Setelah Merge)   : {rows_initial}")
    print(f"Total Baris Akhir (Setelah Clean)  : {rows_final}")
    print(f"Jumlah Duplikat yang Dihapus       : {rows_initial - rows_final}")
    print("-" * 50)
    print("\nInformasi Struktur df_master:")
```

```python
# --- TAHAP 3: DATA WRANGLING & MANIPULASI ---
print("\nMemulai Tahap 3: Data Wrangling & Manipulasi...\n")

# 1. Filtering Data Sukses
# Kita hanya fokus pada pesanan yang tidak batal
df_sukses = df_master[df_master['Status Pesanan'].isin(['Selesai', 'Telah Dikirim'])].copy()

# ---------------------------------------------------------
# EKSPLORASI 1: TOP 10 KATEGORI PRODUK (RANKING)
# ---------------------------------------------------------
top_10_kategori = df_sukses.groupby('product_category')['Jumlah'].sum().sort_values(ascending=False).head(10).reset_index()
top_10_kategori.index = top_10_kategori.index + 1
top_10_kategori.columns = ['Kategori Produk', 'Total Unit Terjual']

print("1. TABEL RANKING: TOP 10 KATEGORI TERLARIS")
display(top_10_kategori)
print("-" * 50)

# ---------------------------------------------------------
# EKSPLORASI 2: SEBARAN PASAR PER PROVINSI (TOP 5)
# ---------------------------------------------------------
top_provinsi = df_sukses.groupby('Provinsi')['order_id'].nunique().sort_values(ascending=False).head(5).reset_index()
top_provinsi.index = top_provinsi.index + 1
top_provinsi.columns = ['Provinsi', 'Jumlah Transaksi']

print("2. TABEL DEMOGRAFI: 5 PROVINSI DENGAN TRANSAKSI TERTINGGI")
display(top_provinsi)
print("-" * 50)

# ---------------------------------------------------------
# EKSPLORASI 3: ANALISIS METODE PEMBAYARAN (DALAM %)
# ---------------------------------------------------------
metode_bayar = df_sukses['Metode Pembayaran'].value_counts(normalize=True).reset_index()
metode_bayar.columns = ['Metode Pembayaran', 'Persentase']
metode_bayar['Persentase'] = (metode_bayar['Persentase'] * 100).round(2).astype(str) + '%'

print("3. TABEL PERILAKU: PREFERENSI METODE PEMBAYARAN")
display(metode_bayar)
print("-" * 50)

# ---------------------------------------------------------
# EKSPLORASI 4: ANALISIS WAKTU (KAPAN PEMBELI BELANJA?)
# ---------------------------------------------------------
# Mengambil komponen jam dari kolom datetime
df_sukses['Jam'] = df_sukses['Waktu Pesanan Dibuat'].dt.hour
jam_sibuk = df_sukses.groupby('Jam')['order_id'].nunique().sort_values(ascending=False).head(5).reset_index()
jam_sibuk.columns = ['Jam Operasional (WIB)', 'Jumlah Pesanan']

print("4. TABEL WAKTU: 5 JAM PALING SIBUK (GOLDEN HOURS)")
display(jam_sibuk)
print("=" * 50)
```
